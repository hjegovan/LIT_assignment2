```python
# The function is to compare two columns and find the differences and similarities
def col_compare(col_a,col_b):
    a_items = col_a.unique()
    b_items = col_b.unique()
    print("Unique items in A: {}\nUnique items in B: {}\n ".format(len(a_items),len(b_items)))
    # find the differences
    all_dif = list(set(a_items).symmetric_difference(set(b_items)))
    # find the similarities
    overlap = set.intersection(set(a_items), set(b_items))
    print("Difference Between A&B: {}\nSimilarities between A&B: {}\n ".format(len(all_dif),len(overlap)))
    return all_dif,overlap
    
```
